Вариант №13:

$x = \sin(t) * (e^{\cos(t)} - 2\cos(4t) + \sin^5(\frac{1}{12t})),
y = \cos(t) * (e^{\cos(t)} - 2\cos(4t) + \sin^5(\frac{1}{12t})),
t \in [0; 12\pi]$


# РЕШЕНИЕ

```python
import matplotlib.pyplot as plt 
import numpy as np
import math
from math import  sin, cos

t_list = np.arange(start=0, stop=12 * math.pi, step=0.001)
x_list = [sin(t) * ((math.e ** cos(t)) - 2 * cos(4 * t) + sin(1 / 12 * t) ** 5) for t in t_list]
y_list = [cos(t) * ((math.e ** cos(t)) - 2 * cos(4 * t) + sin(1 / 12 * t) ** 5) for t in t_list]


plt.subplot(2,1,1)
plt.plot(t_list, x_list, "-r")
plt.title("2sin³(t)")
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(3,1,3)
plt.plot(t_list, y_list, "-g")
plt.title("2cos³(t)")
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")

plt.show()


plt.plot(x_list, y_list, "-b",)
plt.title("График")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend(["2sin³(t)", ""])
plt.show()
```
