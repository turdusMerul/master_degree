Вариант №4:

$x = 2\sin^3(t),
y = 2\cos^3(t),
t \in [0; 2\pi]$


# РЕШЕНИЕ

```python
import matplotlib.pyplot as plt 
import numpy as np
import math
from math import  sin, cos

t_list = np.arange(start=0, stop=2 * math.pi, step=0.001)
x_list = [2 * sin(t)**3 for t in t_list]
y_list = [2 * cos(t)**3 for t in t_list]


plt.subplot(2,1,1)
plt.plot(t_list, x_list, "-r")
plt.title("2sin³(t)")
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(3,1,3)
plt.plot(t_list, y_list, "-g")
plt.title("2cos³(t)")
plt.grid(True)
plt.xlabel("X")
plt.ylabel("Y")

plt.show()


plt.plot(x_list, y_list, "-b",)
plt.title("График")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend(["2sin³(t)", ""])
plt.show()
```
