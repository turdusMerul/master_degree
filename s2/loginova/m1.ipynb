```python
def validate_input(fn):
    print(fn)


@validate_input
def test_fn():
    print("test")



def get_input(_err_msg: str = None) -> str:
    '''Рекурсивная функция для валидации пользовательского ввода.
    Принимает на вход строку с описанием ошибки, по умолчанию равно None.
    Проверяет является ли пользовательский ввод целым числом,
    если нет - вызывает себя, передавая текст ошибки о несоответствии введённых данных.
    Возвращает введённое целое число в формате строки.'''

    # присвоение переменной inputed пользовательского ввода с текстом ошибки, если она не None 
    __inputed = input(f"{_err_msg + ' ' if _err_msg else ''}Введите целое число: ")
    
    if not __inputed.isnumeric() or ("." in __inputed or "," in __inputed):
        # если пользовательский ввод не похож на числовое значение,
        # вызывается эта же функция с передачей в неё текста ошибки,
        # а результат её выполнения переприсвается переменной inputed
        __inputed = get_input("Введённые данные не соответствуют требованиям!")
    
    return __inputed


def print_is_num_in_range(_number: int, _range: list) -> None:
    '''Выводит в консоль сообщение с информацией о том,
    находится ли введённое пользователем число в диапозоне от -1 до 17'''
    print(f"{_number}{' ' if _number in _range else ' НЕ '}находится в диапозоне от -1 до 17")


if __name__ == "__main__":
    # если данный модуль является основным, то будут вызваны тесты
    for i in range(-5, 20):
        print_is_num_in_range(i, range(-1, 17))
```
